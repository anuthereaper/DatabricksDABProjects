```python
%sql
INSERT INTO dev_etl.gold.agg_age
SELECT 
    CASE 
        WHEN dst_age BETWEEN 0 AND 20 THEN '0-20'
        WHEN dst_age BETWEEN 21 AND 40 THEN '21-40'
        WHEN dst_age BETWEEN 41 AND 60 THEN '41-60'
        WHEN dst_age > 60 THEN '61+'
        ELSE 'Unknown'
    END AS dga_age_bracket,
    COUNT(*) AS dga_count_per_bracket
FROM dev_etl.silver.test
GROUP BY 
    CASE 
        WHEN dst_age BETWEEN 0 AND 20 THEN '0-20'
        WHEN dst_age BETWEEN 21 AND 40 THEN '21-40'
        WHEN dst_age BETWEEN 41 AND 60 THEN '41-60'
        WHEN dst_age > 60 THEN '61+'
        ELSE 'Unknown'
    END
ORDER BY dga_age_bracket;
```
